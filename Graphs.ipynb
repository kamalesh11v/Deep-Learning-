```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, precision_recall_curve, auc

# Simulated data for graphs
np.random.seed(42)  # For reproducibility

# Simulate binary classification predictions and actual labels
y_true = np.random.randint(0, 2, 100)  # Actual labels (0 or 1)
y_pred_prob = np.random.rand(100)  # Predicted probabilities
y_pred = (y_pred_prob >= 0.5).astype(int)  # Predicted labels using a threshold of 0.5

# 1. Confusion Matrix
conf_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6, 5))
plt.imshow(conf_matrix, cmap="Blues", interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()
plt.xticks([0, 1], ["Negative", "Positive"])
plt.yticks([0, 1], ["Negative", "Positive"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(j, i, conf_matrix[i, j], ha="center", va="center", color="black")
plt.show()

# 2. ROC Curve
fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})", color="blue")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid()
plt.show()

# 3. Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_true, y_pred_prob)
plt.figure(figsize=(8, 5))
plt.plot(recall, precision, label="Precision-Recall Curve", color="green")
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid()
plt.legend()
plt.show()

# 4. Line Graph
epochs = np.arange(1, 21)  # 20 epochs
accuracy = 1 - np.exp(-0.2 * epochs)  # Simulated accuracy growth
plt.figure(figsize=(8, 5))
plt.plot(epochs, accuracy * 100, marker='o', label='Training Accuracy')
plt.title('Training Accuracy vs. Epochs (Line Graph)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.grid()
plt.legend()
plt.show()

# 5. Scatter Plot
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)
plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.7, color="purple")
plt.title("Scatter Plot")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.grid()
plt.show()

# 6. Histogram
data = np.random.normal(0, 1, 1000)  # Simulated data
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color="orange", edgecolor="black", alpha=0.7)
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# 7. Diverging Bar Chart
categories = ["A", "B", "C", "D", "E"]
values = [10, -15, 5, -10, 20]
colors = ["green" if v > 0 else "red" for v in values]
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=colors)
plt.axhline(0, color="black", linewidth=1)
plt.title("Diverging Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# 8. Waterfall Chart
components = [100, -20, 30, -10, 50]
cumulative = np.cumsum([0] + components)
x_labels = ["Start", "Component 1", "Component 2", "Component 3", "Component 4", "End"]
plt.figure(figsize=(8, 5))
plt.bar(range(len(components) + 1), cumulative, color="skyblue", edgecolor="black")
plt.title("Waterfall Chart")
plt.xlabel("Stages")
plt.ylabel("Values")
plt.xticks(range(len(x_labels)), x_labels, rotation=45)
plt.show()

# 9. Funnel Chart
stages = ["Visitors", "Sign-ups", "Purchases", "Referrals"]
values = [1000, 700, 500, 300]
plt.figure(figsize=(8, 5))
plt.barh(stages, values, color="lightgreen")
plt.title("Funnel Chart")
plt.xlabel("Count")
plt.ylabel("Stages")
plt.grid()
plt.show()
```
